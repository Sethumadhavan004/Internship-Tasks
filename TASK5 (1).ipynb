```python
questions=["1.How many people have heart disease and how many people does not have heart disease?",
           "2. People of which sex has most heart disease? ",
           "3. People of which sex has which type of heart disease most? ",
           "4. People with which chest pain are prone to have heart disease? "]

questions
```

```python
#Finding the answer to first question

#How many people have heart disease and how many people does not have heart disease

df.target.value_counts()
```

```python
#Ploting the bar for the question
df.target.value_counts().plot(kind="bar", color=["lightblue","orchid"])
plt.title("Heart Disease Patient Count")
plt.xlabel("1=Heart Disease, 0=No Heart Disease")
plt.ylabel("Number of People")
plt.xticks(rotation=0)
```

```python
#plotting a pie chart
df.target.value_counts().plot(kind="pie",figsize=(8,6))
plt.legend(["Heart disease","No heart disease"])
```

```python
#Sex column
# 0 Means Women
#1 means Men

#Target Column
# 0 means Not disease
# 1 means Disease

#Viewing the count of sex column
df.sex.value_counts()
```

```python
#Ploting a pie chart
df.sex.value_counts().plot(kind="pie", figsize=(8,6))
plt.title("Male Female Ratio")
plt.legend(["Male","Female"])
```

```python
# Lets find the answer for 2nd Question

#2. People of which sex has most heart disease

pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x= "target", data=df,hue="sex")
plt.title("Heart Disease Frequency for Sex")
plt.xlabel("0 =No heart disease, 1=Heart Disease");
```

```python
#Number of male is more than double in our dataset than female

#More than 45% male has heart disease and 75% female has heart disease
```

```python
#3 Question. People of which sex has which type of chest pain?

#counting values for different types of chest pain
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind="bar", color=["salmon", "orange","lightblue","red"])
plt.title("Count of various Chest Pain in BAR Chart")
plt.xlabel("Type of G=Chest pains")
plt.ylabel("Count of people ")
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind="bar", color=["green","Violet","Royalblue","Yellow"])
plt.title("Heart disease Vs Type of Chest pain ")
plt.xlabel("0= Female, 1= Male")
plt.ylabel("Count of People")
```

```python
# Most of male has Type 0 of the chest pain

# Incase of Female the Type 0 and Type 2 are almost same.
```

```python
# 4th Question. People with which chest pain are prone to have heart disease?

pd.crosstab(df.target, df.cp)
```

```python
sns.countplot(x="cp", data=df, hue="target");
```

```python
# Most of those with type 0 of chest main has least heart attacks

#Those with type 2 has the highest chances of having a heart attack
#All other types has a higher chance of heart attack only.
```
